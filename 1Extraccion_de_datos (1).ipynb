```python
#Filtro por comparacion
filtro6=df[df["PRECIO_UNITARIO"] > 500]
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv("filtro6.csv")
from google.colab import files


files.download("filtro6.csv")
#Se filtran las ventas cuyo precio unitario del producto supere los 500 pesos
```

```python
#Filtro Y
filtro7=df[(df["MARGEN_TOTAL_CALCULADO"] >= 1000) & (df["FECHA_DOC"] >= "2022-08-01")]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("filtro7.csv")
from google.colab import files


files.download("filtro7.csv")
#Se filtran las ventas cuyo margen total sea igual o mayor a 1000 y la fecha de la venta sea apartir del 1 de agosto
```

```python
#Filtro O
filtro8=df[(df["MARGEN_TOTAL_CALCULADO"] >= 1000) | (df["FECHA_DOC"] >= "2022-08-01")]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("filtro8.csv")
from google.colab import files


files.download("filtro8.csv")
#Se filtran las ventas cuyo margen total sea igual o mayor a 1000 o la fecha de la venta sea apartir del 1 de agosto
```

```python
#Filtro not
filtro9=df[~(df["CANT"] <= 1) & ~(df["FECHA_DOC"] <= "2022-07-01")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("filtro9.csv")
from google.colab import files


files.download("filtro9.csv")
#Se filtran las ventas que su cantidad no sea 1 y que su fecha de venta no sea igual o anterior al 1 de julio
```
